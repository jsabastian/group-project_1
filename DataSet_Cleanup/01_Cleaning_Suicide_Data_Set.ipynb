```python
## We are going to merge our two datasets (Happiness and Suicide) on the column 'country'
```

```python
# bring in dependencies
import pandas as pd
import os
import csv
```

```python
# reading in suicide csv and converting to dataframe

suicide = os.path.join("Resources","Suidice_Rate_Data","master.csv")
suicide_df = pd.read_csv(suicide)
suicide_df.head()
```

```python
# filter DataFrame to only show rows from year 2015

suidice = suicide_df.set_index('country')

suidice2 = suidice.loc[suidice['year'] == 2015, :]
```

```python
suidice2.head()
```

```python
# setting up dataframe to be filtered by country, and set to sum so that we can calculate 
# values for a country's total population, suicide number, suicides per 100k population, gdp for year, and gdp per capita
cP = suidice2.groupby(['country']).sum()
suidice2[' gdp_for_year ($) '] = suidice2[' gdp_for_year ($) '].str.replace(',','')
suidice2[' gdp_for_year ($) '] = suidice2[' gdp_for_year ($) '].astype(float)
```

```python
cP
```

```python
## Creating my variables

# Population
countryPOP = cP['population']

# Total Suicide Number
countrySuiNum = cP['suicides_no']

# Suicides per 100k Population
countrySuiPer_100k = cP['suicides/100k pop']

# Individual Countries GDP for the year 2015
cP2 = suidice2.groupby('country').mean()
countryGDP_1year = cP2[' gdp_for_year ($) ']

# Individual Countries GDP per Capita
countryGDP_pCap = cP2['gdp_per_capita ($)']
```

```python
# Creating a Dictionary with this information zipped together
Suicide_Rates = {
    "Population":countryPOP,
    "Total Number of Suicides":countrySuiNum,
    "Number of Suicides per 100K Population":countrySuiPer_100k,
    "GDP for 2015":countryGDP_1year,
    "GDP per Capita":countryGDP_pCap.astype(int)
}
```

```python
# Make a dataframe out of that dictionary
Suicide_Rates_Data = pd.DataFrame(Suicide_Rates)
Suicide_Rates_Data
```

```python
Suicide_Rates_Data.to_csv("Resources/Suidice_Rate_Data/cleaned_data.csv")
```
