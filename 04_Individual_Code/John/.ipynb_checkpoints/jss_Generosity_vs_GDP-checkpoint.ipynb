```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from sklearn.metrics import r2_score
```

```python
#pandas read in the file
csv = pd.read_csv('merged_data_2.csv')

#check the file output
csv.head()
```

## Comparing 'Generosity' to suicides from the Suicide report

```python
#create the dataframe for the variables of interest
df= csv[['country','Generosity','economy (GDP per Capita)','Number of Suicides per 100K Population']]
```

```python
#assign the variables
x = df['Generosity']
y = df['Number of Suicides per 100K Population']
z = df['economy (GDP per Capita)']

#plot formatting
plt.figure(figsize=(8,5))
plt.scatter(x,y)
plt.title('Generosity of a Nation Vs # of Suicide per capita', fontsize=18)
plt.xlabel('Generosity', fontsize=14)
plt.ylabel('Suicide per capita', fontsize=14)
plt.grid(True)

#trend line variables
l = np.polyfit(x,y,1)
p = np.poly1d(l)(x)

#pearson coefficient and pvalue
pval = stats.pearsonr(x,y)

#rounding the tuple values into a list
tup_round = (round(pval[0], 3), round(pval[1], 3))

plt.plot(x,p,'r')

# adding r-square analysis 
text = f"R^2 = {r2_score(y,p):0.3f}\n Pearson Coeff and p-val:\n {tup_round}"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=14, verticalalignment='top')

plt.savefig('fig1.png')
```

```python
x = df['Generosity']
y = df['Number of Suicides per 100K Population']
z = df['economy (GDP per Capita)']

plt.figure(figsize=(8,5))
plt.scatter(z,y)
plt.title('Economic Outlook Vs # of Suicide per capita', fontsize=18)
plt.xlabel('Economic Outlook (GDP per capita)', fontsize=14)
plt.ylabel('Suicide per capita', fontsize=14)
plt.grid(True)

l = np.polyfit(z,y,1)
p = np.poly1d(l)(z)

pval = stats.pearsonr(z,y)

tup_round = (round(pval[0], 3), round(pval[1], 3))

plt.plot(z,p,'r')

# adding r-square analysis
text = f"R^2 = {r2_score(y,p):0.3f}\n Pearson Coeff and p-val:\n {tup_round}"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=14, verticalalignment='top')

plt.savefig('fig2.png')
```

## Comparing 'Generosity' with 'GDP for 2015' data from the suicide report

```python
x = df['Generosity']
z = df['Number of Suicides per 100K Population']
y = df['economy (GDP per Capita)']

# create a variable for the # of size bins to use as a legend
M = 4

#creating the legend from plot variables
df_1 = pd.DataFrame(data=dict(x=x, y=y, z=z))
bins = np.linspace(df_1.z.min(), df_1.z.max(), M)
grouped = df_1.groupby(np.digitize(df_1.z, bins))

#establish size and labels for the legend
sizes = [50*(i+1.) for i in range(M)]
labels = ['Lowest Suicide Rate per capita', '', '', 'Highest Suicide Rate per capita']
plt.figure(figsize=(10,10))
for i, (name, group) in enumerate(grouped):
    plt.scatter(group.x, group.y, s=sizes[i], alpha=0.5, label=labels[i])
  

plt.title('Generosity of a Nation Vs Economic Outlook', fontsize=18)
plt.xlabel('Generosity', fontsize=14)
plt.ylabel('Economic Outlook (GDP per capita)', fontsize=14)
plt.grid(True)
plt.legend()

l = np.polyfit(x,y,1)
p = np.poly1d(l)(x)

pval = stats.pearsonr(x,y)

tup_round = (round(pval[0], 3), round(pval[1], 3))

plt.plot(x,p,'r')

# adding r-square analysis
text = f"R^2 = {r2_score(y,p):0.3f}\n Pearson Coeff and p-val:\n {tup_round}"
plt.gca().text(0.05, 0.95, text,transform=plt.gca().transAxes,
     fontsize=14, verticalalignment='top')

plt.savefig('fig3.png')

plt.show()
```
